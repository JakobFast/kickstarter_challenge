Column meaning:
 * backers_count: number of supporters, continous
 * blurb: description, string
 * category: json_column (to be treated)
 * converted_pledged_amount: pledged amount in US-Dollar, continous
 * country : country of the project, string
 * created_at : Starting date of project, int, timecode
 * creator : Name of company, creator, string
 * currency : Original currency of project, string
 * currency_symbol : Currency_symbol, string
 * currency_trailing_code : When or not the currency has a trailing code. If “True”, then 1, if “False”, then 0., boolean
 * current_currency :  Actual currency, string
 * deadline: deadline, int, timecode
 * disable comunication : Whether communication for the project was disabled (0 = no, 1 = yes), boolean
 * friends : unclear (null or empty)
 * fx_rate : foreign exchange rate, float
 * goal : amount of money project wants to achieve, float
 * id : id
 * is_backing : unclear (null or false)
 * is_starrable : boolean value if project could potentially be starred
 * is_starred: whether or not a project has been starred (liked and saved) by users
 * permissions: not clear, but only has nan values or empty lists
 * location: location, some missing data, interesting json strings
* pledged - amount pledged in the current_currency, float
* profile : profile information, json
* slug : name of the project with hyphens instead of spaces, str
* source_url - url for the project's category, link, str
* spotlight after a project has been successful, it is spotlighted on the Kickstarter website
* staff_pick: whether a project was highlighted as a staff_pick when it was launched/live
* state - whether a project was successful, failed, canceled, suspending or still live
* state_changed_at - date and time of when a project's status was changed (same as the deadline for successful and failed projects)
* static_usd_rate - conversion rate between the original currency and USD
* urls - url to the project's page
* usd_pledged - amount pledged in USD
* usd_type: domestic or international, string

```python
def column_drop(df):
    columns_to_drop = ['blurb', 'converted_pledged_amount', 'currency_symbol', 'currency_trailing_code', 'fx_rate', 'friends', 'id', 'is_backing', 'is_starred', 'permissions', 'static_usd_rate', 'slug', 'source_url', 'usd_pledged']
    df = df.drop(columns_to_drop, axis = 1)
    return df
```

```python
column_drop(df)
```

```python
def target_value(df):
    df['success'] = df.apply(lambda row:1 if row['pledged']>= row['goal'] else 0, axis = 1)
    return df
```
